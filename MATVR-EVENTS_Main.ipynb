```python
df = pd.read_csv(r'D:\LAB Test\NOKIA POLAND\VIAVI-1044\20240124_111904.csv', skiprows=1, delimiter=';')
```

```python
Events_df = df[["Time","ms counter","Events"]]
```

```python
Events_df["Events"] = Events_df["Events"].astype(str).replace({"nan": None, "None": None, "": None})
```

```python
# Remove rows where "Events" is None
Events_df = Events_df.dropna(subset=["Events"])
```

```python
Events_df
```

```python
print("Unique values in Events column:", Events_df["Events"].unique())
```

```python
# Elimina valore NA colummna Data Stream UP
df = df.dropna(subset=[df.columns[4]])
df = df.dropna(subset=[df.columns[5]])
```

```python
df
```

```python
hex_column = df.columns[5]
df['HexPairCount'] = df[hex_column].apply(lambda x: len(str(x).split()) if pd.notna(x) else 0)
```

```python
df
```

```python
df = df[df['HexPairCount'] == 28]
```

```python
df
```

```python
column_ms_counter = df.columns[2]
```

```python
df = df.drop_duplicates(subset=[column_ms_counter], keep='first')
```

```python
df
```

```python
# Funzione per estrarre gli ottetti 22, 23 e 24
def extract_octets(data_stream):
    data_stream2str = str(data_stream).split()
    # Ci assicuriamo che la lunghezza della stringa sia sufficiente
    if len(data_stream2str) == 28:
        octet_22 = bin(int(data_stream2str[22], 16))[2:].zfill(8)  # L'ottetto 22 si trova nella posizione 21 (indice base 0)
        octet_23 = bin(int(data_stream2str[23], 16))[2:].zfill(8)  # L'ottetto 23 si trova nella posizione 22
        octet_24 = bin(int(data_stream2str[24], 16))[2:].zfill(8)  # L'ottetto 24 si trova nella posizione 23
        return octet_22, octet_23, octet_24
    else:
        return None, None, None  # Ritorna None 

# Applicare la funzione alla colonna 4
df[['octet_22','octet_23','octet_24']] = df.iloc[:, 5].apply(lambda x: pd.Series(extract_octets(x)))
```

```python
df
```

```python
octet_22_bin = df.columns[11]
df['octet_22_SN_6bit'] = df[octet_22_bin].apply(lambda x: x[1:-1] if isinstance(x, str) else x)
```

```python
octet_23_bin = df.columns[12]
df['octet_23_SN_6bit'] = df[octet_23_bin].apply(lambda x: x[1:-1] if isinstance(x, str) else x)
```

```python
octet_24_bin = df.columns[13]
df['octet_24_SN_2bit'] = df[octet_24_bin].apply(lambda x: x[1:-5] if isinstance(x, str) else x)
```

```python
df
```

```python
df['JoinedBinary'] = df.apply(lambda row: 
                              (str(row[df.columns[16]]) if pd.notna(row[df.columns[16]]) else '') + 
                              (str(row[df.columns[15]]) if pd.notna(row[df.columns[15]]) else '') + 
                              (str(row[df.columns[14]]) if pd.notna(row[df.columns[14]]) else ''), axis=1)
```

```python
df
```

```python
df['Decimal'] = df['JoinedBinary'].apply(lambda x: int(x, 2) if x != '' else None)
```

```python
df
```

```python
Decimal_Check = df.columns[18]
df = df.drop_duplicates(subset=[Decimal_Check], keep='first')
```
